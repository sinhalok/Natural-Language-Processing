```python
d = {'articles':[],'links':[]}
for link in soup.find_all('a'):
    if len(link.text.strip())>1 and bool(
        re.match('What Dreams May Come|Train to Nowhere|Six Years And Counting…|Getting Saucy About Food',link.text)):
        
        d['articles'].append(link.text)
        d['links'].append(link['href'])
        print(link.text)
```

```python
articles = pd.DataFrame(d)
```

```python
data = {'article':[], 'link':[], 'text':[]}
for rownum, row in articles.iterrows():
    article = row['articles']
    link = row['links']

    html = requests.get(link).text
    soup = bs(html, 'html.parser')
    
    text_all = soup.find_all('p')[:-2] #Selecting all the paragraphs
    text_data = ""
    
    for text in text_all:
        text_data = text_data +" " +text.text.strip()
        
    if text_data != "":
        #print(text_data)
        data['article'].append(article)
       # data['title'].append(title)
        data['link'].append(link)
        data['text'].append(text_data)
```

```python
def expand_contractions(text):
    for word in text.split():
        if word.lower() in contractions:
            text = text.replace(word, contractions[word.lower()])
    return text
```

```python
data['text'] = [expand_contractions(re.sub('(”|’|“)',"'", text)) for text in data['text']]
data['text'] = [re.sub("—"," ",text) for text in data['text']]
```

```python
def scrub_words(text):
    #Replace \xao characters in text
    text = re.sub('\xa0', ' ', text)
    
    #Replace non ascii / not words and digits
    text = re.sub("(\\W|\\d)",' ',text)
    
    text = re.sub(r"\b[a-zA-Z]\b", '', text)
    
    text = re.sub("—"," ",text)
    
    #Remove extra spaces from the text
    text = re.sub("\s+", ' ', text)
    return text
```

```python
data['text'] = [scrub_words(text) for text in data['text']]
data['text'][0]
```

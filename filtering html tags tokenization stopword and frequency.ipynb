```python
from bs4 import BeautifulSoup
html = page.text
soup=BeautifulSoup(html,'html.parser')
html_tag=[tag.name for tag in soup.find_all()]
print(html_tag)
print(set(html_tag))
```

```python
#filtering html tags 
filter_residue= [s.extract() for s in soup(['style','script','[document]','head','title'])]
text= soup.get_text()
print(text)
```

```python
#tokenization
import nltk
from nltk.tokenize import word_tokenize
tokens=word_tokenize(text)
print(tokens)
```

```python
#stopword removal
from nltk.corpus import stopwords
stopwords=list(nltk.corpus.stopwords.words('english'))
words=[word for word in tokens if word not in stopwords]
print(words)
```

```python
#display tokens and frequency of each words
fdist=nltk.FreqDist(words)
for word,frequency in fdist.most_common():
    print('{} ==> {}'.format(word,frequency))
```
